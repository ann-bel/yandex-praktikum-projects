````python
#–≤—ã–≤–µ–¥–µ–º —Ç–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö –ø–æ —Å—Ç–æ–ª–±—Ü–∞–º
data.info()
````

–ù–µ–æ–±—Ö–æ–¥–∏–º–æ –∏–∑–º–µ–Ω–∏—Ç—å —Ç–∏–ø –¥–∞–Ω—ã—Ö –≤ —Å—Ç–æ–ª–±—Ü–µ —Å –æ—Ü–µ–Ω–∫–∞–º–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏ –≥–æ–¥–æ–º –≤—ã–ø—É—Å–∫–∞.

````python
#–∑–∞–º–µ–Ω–∏–º tbd –≤ —Å—Ç–æ–ª–±—Ü–µ user_score –Ω–∞ NaN
data['user_score'] = data['user_score'].replace('tbd',np.NaN)
````

````python
#—á—Ç–æ–±—ã —Å—ç–∫–æ–Ω–æ–º–∏—Ç—å –ø–∞–º—è—Ç—å –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –¥–∞–Ω–Ω—ã–µ –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö —Å float64 –Ω–∞ float32
columns = ['na_sales',
    'eu_sales',
    'jp_sales',
    'other_sales',
    'critic_score',
    'user_score'
]

for column in columns:
    data[column] = data[column].astype('float32')
````

````python
#–ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –¥–∞–Ω–Ω—ã–µ –≤ —Å—Ç–æ–ª–±—Ü–µ –≥–æ–¥–∞ —Ä–µ–ª–∏–∑–∞ –Ω–∞ —Ü–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω—ã–µ
data['year_of_release'] = np.floor(pd.to_numeric(data['year_of_release'], 
    errors='coerce')).astype('Int32')
````

````python
# –≤–Ω–æ–≤—å –ø—Ä–æ–≤–µ—Ä–∏–º —Ç–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö
data.info()
````

````python
# –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞
temp = data.copy() 
time_life = temp.pivot_table(index='platform', values='year_of_release', aggfunc=['min', 'max']).reset_index()
time_life['life'] = time_life['max'] - time_life['min'] + 1 # —Å—Ä–æ–∫ –∂–∏–∑–Ω–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã, –¥–æ–ª–∂–Ω—ã –≤—Ö–æ–¥–∏—Ç—å –≥—Ä–∞–Ω–∏—Ü—ã –¥–∏–∞–ø–∞–∑–æ–Ω–∞, –ø–æ—ç—Ç–æ–º—É +1
time_life['life'].median()
````

````python
#—Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –≤ –æ—Ç–¥–µ–ª—å–Ω—ã–π –î–§ –ø—Ä–æ–¥–∞–∂ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º —Å 2012 –≥–æ–¥–∞
data_selection = data.query('year_of_release >= 2012').reset_index(drop=True)
data_selection.head()
````

````python
(data_selection
    .pivot_table(index='year_of_release',
                            columns='platform',
                            values='sum_sales', 
                            aggfunc='sum')
    .plot(grid=True, figsize=(15,7))
);
plt.xlabel('–ì–æ–¥ —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏');
plt.ylabel('–ü—Ä–æ–¥–∞–∂–∏ (–º–ª–Ω. —à—Ç—É–∫)');
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—É–º–º–∞—Ä–Ω—ã—Ö –ø—Ä–æ–¥–∞–∂ –ø–ª–∞—Ç—Ñ–æ—Ä–º –ø–æ –≥–æ–¥–∞–º');
plt.xticks([2012,2013,2014,2015,2016]);
````

````python
#–ø–æ—Å—á–∏—Ç–∞–µ–º –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—é –º–µ–∂–¥—É –æ—Ü–µ–Ω–∫–æ–π –∫—Ä–∏—Ç–∏–∫–æ–≤ –∏ –¥–æ—Ö–æ–¥–∞–º–∏ –æ—Ç –∏–≥—Ä
data_selection_X360_corr = data_selection_X360['critic_score'].corr(data_selection_X360['sum_sales'])

print(f'–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è: {data_selection_X360_corr:.2f}')

#–ø–æ—Å—Ç—Ä–æ–∏–º –¥–∏–∞–≥—Ä–∞–º–º—É —Ä–∞—Å—Å–µ—è–Ω–∏—è
data_selection_X360.plot(kind='scatter',
    x='critic_score',
    y='sum_sales',
    alpha=0.5,
    figsize=(10,7),
    grid=True
);
plt.xlabel('–û—Ü–µ–Ω–∫–∞ –∫—Ä–∏—Ç–∏–∫–æ–≤');
plt.ylabel('–ü—Ä–æ–¥–∞–∂–∏ (–º–ª–Ω. —à—Ç—É–∫)');
plt.title('–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å —Å—É–º–º–∞—Ä–Ω—ã—Ö –ø—Ä–æ–¥–∞–∂ –∏–≥—Ä—ã –æ—Ç –æ—Ü–µ–Ω–∫–∏ –∫—Ä–∏—Ç–∏–∫–æ–≤');
````

–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–µ–∂–¥—É –æ—Ü–µ–Ω–∫–æ–π –∫—Ä–∏—Ç–∏–∫–æ–≤ –∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–∞—è, –Ω–æ –Ω–µ –æ—á–µ–Ω—å —Å–∏–ª—å–Ω–∞—è. –° —Ä–æ—Å—Ç–æ–º –æ—Ü–µ–Ω–∫–∏, —Ä–∞—Å—Ç—É—Ç –∏ –ø—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä. 

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –∫–∞–∫ –≤–ª–∏—è—é—Ç –æ—Ç–∑—ã–≤—ã –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏

````python
data_selection_X360_corr_users = data_selection_X360['user_score'].corr(data_selection_X360['sum_sales'])

print(f'–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è: {data_selection_X360_corr_users:.2f}')

print(data_selection_X360.plot(kind='scatter',
    x='user_score',
    y='sum_sales',
    alpha=0.5,
    figsize=(8,5),
    grid=True
))
plt.title('–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å —Å—É–º–º–∞—Ä–Ω—ã—Ö –ø—Ä–æ–¥–∞–∂ –∏–≥—Ä—ã –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–π –æ—Ü–µ–Ω–∫–∏',fontsize=10, fontweight="bold");
plt.xlabel('–û—Ü–µ–Ω–∫–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π',fontsize=10, fontweight="bold");
plt.ylabel('–°—É–º–º–∞—Ä–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏',fontsize=10, fontweight="bold");
````

–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–∞—è, –æ–¥–Ω–∞–∫–æ –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏ –Ω–µ –æ–∫–∞–∑—ã–≤–∞–µ—Ç –≤–ª–∏—è–Ω–∏—è –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä.

–ü–æ—Å—á–∏—Ç–∞–µ–º –¥–ª—è —Ç–æ–ø–æ–≤—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—é –º–µ–∂–¥—É –ø—Ä–æ–¥–∞–∂–∞–º–∏ –∏–≥—Ä –∏ –æ—Ü–µ–Ω–∫–æ–π –∫—Ä–∏—Ç–∏–∫–æ–≤ –∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.

````python
#–Ω–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏: –ø–∞—Ç—Ñ–æ—Ä–º–∞ –∏ –æ—Ü–µ–Ω–∫–∞
def print_correlation_and_plot(platform_choose, score):
    selected_platform = data_selection.query('platform == @platform_choose')
    selected_platform_corr = selected_platform[score].corr(selected_platform['sum_sales'])
    selected_platform_corr = round(selected_platform_corr,2)
    selected_platform.plot(kind='scatter',
                          x=score,
                          y='sum_sales',
                          alpha=0.5,
                          figsize=(10,5),
                          grid=True
                         )
    plt.title('–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å —Å—É–º–º–∞—Ä–Ω—ã—Ö –ø—Ä–æ–¥–∞–∂ –∏–≥—Ä –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ '+platform_choose+' –æ—Ç '+score, fontsize=10, fontweight="bold");
    plt.xlabel('–û—Ü–µ–Ω–∫–∞',fontsize=10, fontweight="bold");
    plt.ylabel('–°—É–º–º–∞—Ä–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏',fontsize=10, fontweight="bold");
    print('–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è:'+' '+str(selected_platform_corr))
    
````

````python
#–≤—ã–±–µ—Ä–µ–º –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º–∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏ –¥–ª—è –ø–µ—Ä–∏–æ–¥–∞ —Å 2012 –ø–æ 2016 –≥–æ–¥–∞
top_platform = (data_selection
    .groupby(['platform'])
    .agg(sales=('sum_sales','sum'))
    .sort_values(by='sales', ascending=False)
    .reset_index()
)
top_platform = top_platform[:5]
top_platform.columns = ['–ò–≥—Ä–æ–≤–∞—è –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞','–°—É–º–º–∞—Ä–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏']
top_platform
````

````python
top_platform['–ò–≥—Ä–æ–≤–∞—è –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞'].unique().tolist()
````

````python
#–∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É –ø—Ä–æ–¥–∞–∂–∞–º–∏ –∏–≥—Ä –∏ –æ—Ü–µ–Ω–∫–æ–π –∫—Ä–∏—Ç–∏–∫–æ–≤ –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ PS4
print_correlation_and_plot('PS4','critic_score')
````

````python
#–∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É –ø—Ä–æ–¥–∞–∂–∞–º–∏ –∏–≥—Ä –∏ –æ—Ü–µ–Ω–∫–æ–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ PS4
print_correlation_and_plot('PS4','user_score')
````

````python
#–∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É –ø—Ä–æ–¥–∞–∂–∞–º–∏ –∏–≥—Ä –∏ –æ—Ü–µ–Ω–∫–æ–π –∫—Ä–∏—Ç–∏–∫–æ–≤ –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ PS3
print_correlation_and_plot('PS3','critic_score')
````

````python
#–∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É –ø—Ä–æ–¥–∞–∂–∞–º–∏ –∏–≥—Ä –∏ –æ—Ü–µ–Ω–∫–æ–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ PS3
print_correlation_and_plot('PS3','user_score')
````

````python
#–∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É –ø—Ä–æ–¥–∞–∂–∞–º–∏ –∏–≥—Ä –∏ –æ—Ü–µ–Ω–∫–æ–π –∫—Ä–∏—Ç–∏–∫–æ–≤ –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ DS3
print_correlation_and_plot('3DS','critic_score')
````

````python
#–∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É –ø—Ä–æ–¥–∞–∂–∞–º–∏ –∏–≥—Ä –∏ –æ—Ü–µ–Ω–∫–æ–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ DS3
print_correlation_and_plot('3DS','user_score')
````

````python
#–∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É –ø—Ä–æ–¥–∞–∂–∞–º–∏ –∏–≥—Ä –∏ –æ—Ü–µ–Ω–∫–æ–π –∫—Ä–∏—Ç–∏–∫–æ–≤ –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ XOne
print_correlation_and_plot('XOne','critic_score')
````

````python
#–∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É –ø—Ä–æ–¥–∞–∂–∞–º–∏ –∏–≥—Ä –∏ –æ—Ü–µ–Ω–∫–æ–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ XOne
print_correlation_and_plot('XOne','user_score')
````

### –ü–æ—Ä—Ç—Ä–µ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ —Ä–µ–≥–∏–æ–Ω–∞–º

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –¥–∞–Ω–Ω—ã–µ –ø–æ –∏–≥—Ä–∞–º –≤ —Ä–∞–∑—Ä–µ–∑–µ —Ä–µ–≥–∏–æ–Ω–æ–≤.

#### –¢–æ–ø-5 –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º –ø–æ —Ä–µ–≥–∏–æ–Ω–∞–º

````python
#—Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –≤—ã–≤–æ–¥–∞ —Å—Ç–æ–ª–±—á–∞—Ç–æ–π –¥–∏–∞–≥—Ä–∞–º–º—ã
def print_top_5_bar(group_option, region, ax,  name, quantity=5):
    pivot = data_selection.groupby(group_option).agg({region:'sum'}).sort_values(by=region, ascending=False)[:quantity].reset_index()
    other_option = data_selection.groupby(group_option).agg({region:'sum'}).sort_values(by=region, ascending=False)[quantity:][region].sum()
    other_values_line = {group_option:'Other',region:other_option}
    pivot = pivot.append(other_values_line,ignore_index=True)
    pivot_platform = list(pivot[group_option])
    title = {'na_sales':'–ü—Ä–æ–¥–∞–∂–∏ –≤ –°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–µ', 'eu_sales':'–ü—Ä–æ–¥–∞–∂–∏ –≤ –ï–≤—Ä–æ–ø–µ', 'jp_sales':'–ü—Ä–æ–¥–∞–∂–∏ –≤ –Ø–ø–æ–Ω–∏–∏'}
    color = {'na_sales': 'lightblue', 'eu_sales':'orange', 'jp_sales':'green'}
    plot = pivot.plot(kind='bar', title = title[region], ax=axes[ax], fontsize=13, rot=0, color=color[region])
    plot.legend(prop={'size':15})
    plot.set_xlabel(name)
    plot.set_xticklabels(pivot_platform)
    plot.title.set_size(18)
    plot.set_ylabel('–ü—Ä–æ–¥–∞–∂–∏, –º–ª–Ω –∫–æ–ø–∏–π')
````

````python
#—Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –≤—ã–≤–æ–¥–∞ –∫—Ä—É–≥–æ–≤–æ–π –¥–∏–∞–≥—Ä–∞–º–º—ã
def print_top_5_pies(group_option, regions,quantity=5):    
    fig, axes = plt.subplots(nrows=1, ncols=len(regions))    
    i = 0
    for region in regions:
        pivot = data_selection.groupby(group_option).agg({region:'sum'}).sort_values(by=region, ascending=False).reset_index()[:quantity]
        other_option = data_selection.groupby(group_option).agg({region:'sum'}).sort_values(by=region, ascending=False).reset_index()[quantity:][region].sum()
        other_values_line = {group_option:'Other',region:other_option}
        pivot = pivot.append(other_values_line,ignore_index=True)
        pivot_platform = list(pivot[group_option])
        titles = {'na_sales':'–ü—Ä–æ–¥–∞–∂–∏ –≤ –°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–µ', 'eu_sales':'–ü—Ä–æ–¥–∞–∂–∏ –≤ –ï–≤—Ä–æ–ø–µ', 'jp_sales':'–ü—Ä–æ–¥–∞–∂–∏ –≤ –Ø–ø–æ–Ω–∏–∏'}     
        pivot.plot(ax=axes[i], y=region,kind='pie',labels=pivot_platform, autopct='%1.1f%%', fontsize=13,rot=15, figsize=(20,20), legend=False, title=titles[region], startangle=10);
        plt.ylabel("")
      
        i+=1
  
````

````python
fig,axes = plt.subplots(1,3, figsize=(25,6))

print_top_5_bar('platform','na_sales',0,'–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞')
print_top_5_bar('platform','eu_sales',1,'–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞')
print_top_5_bar('platform','jp_sales',2,'–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞')
````

````python
print_top_5_pies('platform',['na_sales', 'eu_sales','jp_sales'])
````

````python
fig,axes = plt.subplots(1,3, figsize=(25,6))

print_top_5_bar('genre','na_sales',0, '–ñ–∞–Ω—Ä')
print_top_5_bar('genre','eu_sales',1, '–ñ–∞–Ω—Ä')
print_top_5_bar('genre','jp_sales',2, '–ñ–∞–Ω—Ä')
````

````python
def print_rating_pies(group_option, regions,quantity=5):    
    fig, axes = plt.subplots(nrows=1, ncols=len(regions))    
    i = 0
    for region in regions:
        pivot = data_selection.groupby(group_option).agg({region:'sum'}).sort_values(by=region, ascending=False).reset_index()[:quantity]
        pivot_platform = list(pivot[group_option])
        titles = {'na_sales':'–ü—Ä–æ–¥–∞–∂–∏ –≤ –°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–µ', 'eu_sales':'–ü—Ä–æ–¥–∞–∂–∏ –≤ –ï–≤—Ä–æ–ø–µ', 'jp_sales':'–ü—Ä–æ–¥–∞–∂–∏ –≤ –Ø–ø–æ–Ω–∏–∏'}     
        pivot.plot(ax=axes[i], y=region,kind='pie',labels=pivot_platform, autopct='%1.1f%%', fontsize=13,rot=15, figsize=(20,20), legend=False, title=titles[region], startangle=10);
        plt.ylabel("")
      
        i+=1
  
````

````python
#–ø–æ—Å–º–æ—Ç–∏–º –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä –≤ —Ä–∞–∑—Ä–µ–∑–µ –≤–æ–∑—Ä–∞—Å—Ç–Ω—ã—Ö –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–π, –≤ –º–ª–Ω –ø—Ä–æ–¥–∞–Ω–Ω—ã—Ö –∫–æ–ø–∏–π
#—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è nr –Ω–µ –±—É–¥–µ–º, —Ç.–∫. —Ä–µ–π—Ç–∏–Ω–≥ ESRB –ø—Ä–∏–º–µ–Ω—è–ª—Å—è —Ç–æ–ª—å–∫–æ –¥–ª—è –∏–≥—Ä, –≤—ã–ø—É—â–µ–Ω–Ω—ã—Ö –≤ –ö–∞–Ω–∞–¥–µ –∏ –°–®–ê
#—Å–æ—Ö—Ä–∞–Ω–∏–º –Ω–∞–∑–≤–∞–Ω–∏–µ —Ä–µ–≥–∏–æ–Ω–æ–≤ –≤ —Å–ø–∏—Å–æ–∫
regions=['na_sales','eu_sales','jp_sales']

#—Å –ø–æ–º–æ—â—å—é —Ü–∏–∫–ª–∞ –≤—ã–≤–µ–¥–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ —Ä–µ–≥–∏–æ–Ω–∞–º
for region in regions:
    rating_by_region = (data_selection
                        .groupby(['rating'])
                        .agg(sales=(region,'sum'))
                        .sort_values(by='sales',ascending=False)
    )                    
    print('–†–µ–≥–∏–æ–Ω –ø—Ä–æ–¥–∞–∂:', region)
    print()
    print(rating_by_region)
    print()
````

````python
#–æ—Ç—Ñ–∏–ª—å—Ç—Ä—É–µ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –ø–æ –æ—Ü–µ–Ω–∫–∞–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ —Ä–∞–∑—Ä–µ–∑–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º, —É–¥–∞–ª–∏–º –ø—Ä–æ–ø—É—Å–∫–∏
data_Xbox = data_selection[data_selection['platform'] == 'XOne']['user_score'].dropna()
data_PC = data_selection[data_selection['platform'] == 'PC']['user_score'].dropna()

#–ø—Ä–æ–≤–µ–¥–µ–º t-—Ç–µ—Å—Ç
results = st.ttest_ind(data_Xbox,data_PC, equal_var = False)

#–∑–∞–¥–∞–µ–º –ø–æ—Ä–æ–≥ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
alpha = 0.05

print('p-–∑–Ω–∞—á–µ–Ω–∏–µ:',results.pvalue)

if results.pvalue < alpha:
    print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
else:
    print('–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
````
