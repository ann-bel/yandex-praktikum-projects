# Изучение закономерностей, определяющих успешность игр

### Задача
Используя исторические данные о продажах компьютерных игр, оценки пользователей и экспертов, жанры и платформы, выявить закономерности, определяющие успешность игры 

# Описание проекта

Выявлены параметры, определяющие успешность игры в разных регионах мира. 

На основании этого подготовлен отчет для магазина компьютерных игр для планирования рекламных кампаний. Проведена предобработка данных, анализ. Выбран актуальный период для анализа. Составлены портреты пользователей каждого региона. 

Проверены гипотезы: 
- средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- средние пользовательские рейтинги жанров Action и Sports разные. 

При анализе использован критерий Стьюдента для независимых выборок.

**Описание данных**
- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

### Импорт основных библиотек и выгрузка дынных 

````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
from scipy import stats as st
import warnings
warnings.simplefilter('ignore')

#это код, который выводит все колонки таблицы
pd.set_option('display.max_columns', None)

#это отображает float с двумя знаками после точки
pd.options.display.float_format = '{:.2f}'.format
````

````python
data = pd.read_csv('https://code.s3.yandex.net/datasets/games.csv')
data.head(10)
````

### Предобработка данных

#### Приведение названий столбцов к единому регистру

Приведем все названия столбцов к нижнему регистру.

````python
data.columns = data.columns.str.lower()
data.head(5)
````

#### Изменение типа данных в столбцах

````python
#выведем типы данных по столбцам
data.info()
````

Необходимо изменить тип даных в столбце с оценками пользователей и годом выпуска.

Изменим тип данных. 

TBD - аббревиатура от английского To Be Determined (будет определено) или To Be Decided (будет решено). Используется, если какая-то информация еще не определена или решение по вопросу не принято. Акроним служит для отметки неясностей или пропусков, которые надо заполнить. Следовательно строчки с такой аббревиатурой можно заменить на NaN, т.к. решение пользователями еще не принято.

````python
#заменим tbd в столбце user_score на NaN
data['user_score'] = data['user_score'].replace('tbd',np.NaN)
````

````python
#чтобы сэкономить память преобразуем данные в столбцах с float64 на float32
columns = ['na_sales',
    'eu_sales',
    'jp_sales',
    'other_sales',
    'critic_score',
    'user_score'
]

for column in columns:
    data[column] = data[column].astype('float32')
````

````python
#преобразуем данные в столбце года релиза на целочисленные
data['year_of_release'] = np.floor(pd.to_numeric(data['year_of_release'], 
    errors='coerce')).astype('Int32')
````

````python
# вновь проверим типы данных в столбцах
data.info()
````

````python
data.head(10)
````

#### Обработка пропусков 

````python
#подсчитаем количество пропусков в данных
data.isna().sum()
````

````python
# проверим, что за данные в строках с пустым значением по столбцу genre, 
#возможно их можно заменить
data[data['genre'].isnull()]
````

Как мы видим данные отсутствуют не только в столбце жанра игры, но и по столбцу "Name", мы не можем адекватно заполнить эти значения и их немного, значит логичнее всего удалить эти строки.

Пропущенные данные в столбце года релиза нельзя адекватно заполнить медианным или средним значением, т.к. таких значений мало, считаем, что их тоже нужно удалить.

````python
data = data.dropna(subset = ['genre','year_of_release'])
data.isna().sum()
````

**Пропуски в значениях столбцов:**
- critic_score
- user_score
- rating

могли возникнуть по следующим причинам:

- ошибка выгрузки данных;
- человечиский фактор;
- игра еще не была оценена.

Удалять или заменять проруски в этих столбцах нельзя, так как это приведет к искажению данных, оставим их без изменений.

#### Обработка неявных дубликатов

Рейтинг «K-A» («Kids to Adults»)— «Для детей и взрослых»: Содержание игры вполне подходит для возрастной категории от 6 лет и старше. Такие игры подходят для всех возрастов. Они могут содержать минимальное насилие или немного грубый разговорный язык. Этот рейтинг был заменён на «Everyone» 1 января 1998 года и может встречаться только в играх, изданных ранее. Поэтому для целей исследования считаем необходимым заменить значения `К-А` на `Е`

````python
# заменим значения в столбце рейтинга с K-A на Е
data['rating'] = data['rating'].replace('K-A','E')
data['rating'].unique()
````

Также заменим NaN в столбце рейтинга на "nr", ведь точно определить рейтинг мы не можем. Пропуски в значениях возможны из-за ошибок выгрузки или из-за отсутствия оценки как таковой. В нашем случае рейтинг ESRB только для США и Канады, поэтому на играх выпушенных в других странах этот ретинг не распростроняется.

Также пропуски в значениях этого солбца могут означать, что игра не оценивалась рейтингом ESRB.

````python
data['rating'] = data['rating'].fillna('nr')
````

````python
data['rating'].value_counts()
````

#### Обработка явных дубликатов

````python
#проверим есть ли явные дубликаты
data.duplicated().sum()
````

Явных дубликатов не оказалось.

#### Подсчет суммарных продаж по всем регионам

````python
#Создадим столбец с суммарными продажами по всем регионам
data['sum_sales'] = data.loc[:,['na_sales','eu_sales', 'jp_sales', 'other_sales']].sum(axis=1)
````

````python
data.head()
````

**Промежуточный вывод**

В ходе предобработки данных, мы:
- выяснили, что не все данные имеют правильный числовой тип, поэтому заменили годы выпуска на целочисленный, а оценки пользователей на значения с плавающей запятой, таже сменили float64 на float32 для экономии памяти;
- выявили явные дубликаты и удалили их;
- заменили устаревший рейтинг `К-А` на `Е` для единообразия оценки;
- создали дополнительный столбец, отражающий суммарные продажи по всем регионам.

Мы видим активный рост выпуска игр с 1994 года, пик релизов приходится на 2008-2009 года ~ 1400 штук. 

Активный рост рынка привлекал в себя все больше инвестиций, деньги получали практически все проекты, люди верили, что смогут не только вернуть свои вложения,но и преумножить их. Это было время экспериментов, проб и ошибок.

Однако, после финансового кризиса 2008 года, из-за лопнувшего пузыря необеспеченных ипотечных займов, который повлиял не только на банковский сектор, но и на многие другие отрасли, оставив, по оценкам экспертов, около 8 млн жителей только в США без работы, игровая индустрия несла коллосальные убытки, фирмы закрывались или сливались с другими корпорациями, продаваясь за символические деньги. В это время инвесторы и производители игр стали намного разборчивее, выпуск игр на PC и консолях сократился. Фокус переместился на мобильные и онлайн-игры.  

С 2014 года идет небольшой рост (в 2016 году мы видим провал, но нам известно. что данные за этот год неполные, поэтому вролне возможен и рост). 

Рассмотрим выпуск игр на определенных платформах за все периоды.

````python
platform_top_sales = (data
    .groupby(['platform'])
    .agg(sales=('sum_sales','sum'))
    .sort_values(by='sales',ascending=False)

)

platform_top_sales.plot(kind='bar',figsize=(15,10));
plt.xlabel('Платформа');
plt.ylabel('Количество выпущенных игр', );
plt.title('Количество выпущенных игр по платформам');
````

Как мы видим из диаграммы наиболее популярными являются платформы: PS, DS,Wii, PS3, X360 и PS2. Посмотрим, как менялись продажи по годам у этих платформ.

````python
#выделим названия 6-ти самых прибыльных платформ
platform_top_sales = platform_top_sales.reset_index()
platform_top_sales = platform_top_sales[:6]['platform'].tolist()
````

````python
platform_sales = data.query("platform == ['PS2', 'X360', 'PS3', 'Wii', 'DS', 'PS']")
````

````python
(platform_sales.pivot_table(index='year_of_release',
                           columns='platform', 
                           values='sum_sales', 
                           aggfunc='sum')
    .plot(grid=True, figsize=(15,7))
);
plt.xlabel('Год реализации');
plt.ylabel('Продажи(млн.)');
plt.title('Распределение суммарных продаж платформ по годам');
````

````python
# посчитаем среднюю продолжительность работы платформ 
temp = data.copy() 
time_life = temp.pivot_table(index='platform', values='year_of_release', aggfunc=['min', 'max']).reset_index()
time_life['life'] = time_life['max'] - time_life['min'] + 1 # срок жизни платформы, должны входить границы диапазона, поэтому +1
time_life['life'].median()
````

Из графика можно сделать следующие выводы: 
- платформы живут в среднем 7 лет; 
- пик продаж приходится на пятый год существования платформы, после этого идет спад;
- с 2004 по 2016 наблюдается увеличените количества платформ;
- новые платформы появляются примерно каждые 5 лет;
- с 2010 года наблюдается спад продаж по все платформам.

Предполагаю, что спад продаж игр на консолях может быть вызван большей доступностью и популярностью мобильных и компьютерных игр. Помимо этого снижение продаж игр на эти приставки можно объяснить выходом новых платформ. 

Считаем, что для составления прогноза по продажам целесообразно взять период с 2012, т.к. за этот период вышли новые приставки и по ним мы будем ожидать пика продаж как раз в 2017-2018 годах. Сохраним эти года в отдельный датафрейм.

````python
#сохранение в отдельный ДФ продаж по платформам с 2012 года
data_selection = data.query('year_of_release >= 2012', engine='python').reset_index(drop=True)
data_selection.head()
````

#### Анализ динамики продаж по платформам с 2012 года и выбор потанциально прибыльных платформ

Посчитаем для топовых платформ корреляцию между продажами игр и оценкой критиков и пользователей.

````python
#напишем функцию для создания графика корреляции с параметрами: патформа и оценка
def print_correlation_and_plot(platform_choose, score):
    selected_platform = data_selection.query('platform == @platform_choose')
    selected_platform_corr = selected_platform[score].corr(selected_platform['sum_sales'])
    selected_platform_corr = round(selected_platform_corr,2)
    selected_platform.plot(kind='scatter',
                          x=score,
                          y='sum_sales',
                          alpha=0.5,
                          figsize=(10,5),
                          grid=True
                         )
    plt.title('Зависимость суммарных продаж игр на платформе '+platform_choose+' от '+score, fontsize=10, fontweight="bold");
    plt.xlabel('Оценка',fontsize=10, fontweight="bold");
    plt.ylabel('Суммарные продажи',fontsize=10, fontweight="bold");
    print('Корреляция:'+' '+str(selected_platform_corr))
    
````

#### Топ-5 популярных платформ по регионам

````python
#функция для вывода столбчатой диаграммы
def print_top_5_bar(group_option, region, ax,  name, quantity=5):
    pivot = data_selection.groupby(group_option).agg({region:'sum'}).sort_values(by=region, ascending=False)[:quantity].reset_index()
    other_option = data_selection.groupby(group_option).agg({region:'sum'}).sort_values(by=region, ascending=False)[quantity:][region].sum()
    other_values_line = {group_option:'Other',region:other_option}
    pivot = pivot.append(other_values_line,ignore_index=True)
    pivot_platform = list(pivot[group_option])
    title = {'na_sales':'Продажи в Северной Америке', 'eu_sales':'Продажи в Европе', 'jp_sales':'Продажи в Японии'}
    color = {'na_sales': 'lightblue', 'eu_sales':'orange', 'jp_sales':'green'}
    plot = pivot.plot(kind='bar', title = title[region], ax=axes[ax], fontsize=13, rot=0, color=color[region])
    plot.legend(prop={'size':15})
    plot.set_xlabel(name)
    plot.set_xticklabels(pivot_platform)
    plot.title.set_size(18)
    plot.set_ylabel('Продажи, млн копий')
````

````python
#функция для вывода круговой диаграммы
def print_top_5_pies(group_option, regions,quantity=5):    
    fig, axes = plt.subplots(nrows=1, ncols=len(regions))    
    i = 0
    for region in regions:
        pivot = data_selection.groupby(group_option).agg({region:'sum'}).sort_values(by=region, ascending=False).reset_index()[:quantity]
        other_option = data_selection.groupby(group_option).agg({region:'sum'}).sort_values(by=region, ascending=False).reset_index()[quantity:][region].sum()
        other_values_line = {group_option:'Other',region:other_option}
        pivot = pivot.append(other_values_line,ignore_index=True)
        pivot_platform = list(pivot[group_option])
        titles = {'na_sales':'Продажи в Северной Америке', 'eu_sales':'Продажи в Европе', 'jp_sales':'Продажи в Японии'}     
        pivot.plot(ax=axes[i], y=region,kind='pie',labels=pivot_platform, autopct='%1.1f%%', fontsize=13,rot=15, figsize=(20,20), legend=False, title=titles[region], startangle=10);
        plt.ylabel("")
      
        i+=1
  
````

````python
fig,axes = plt.subplots(1,3, figsize=(25,6))

print_top_5_bar('platform','na_sales',0,'Платформа')
print_top_5_bar('platform','eu_sales',1,'Платформа')
print_top_5_bar('platform','jp_sales',2,'Платформа')
````

### Проверка  гипотез

Определим принцип формулирования гипотез. 

Нулевая гипотеза(H0) - это основное проверяемое предположение, которое формулируется как отсутствие различий, отсутствие влияние фактора, отсутствие эффекта.

Альтернативная гипотеза(H1) - это гипотеза о значимости различий. 

#### Проверка гипотезы: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

```
H_0: Средний пользовательский рейтинг платформы Xbox One = среднему пользовательскому рейтингу платформы PC`
H_1: Средний пользовательский рейтинг платформы Xbox One ≠ cреднему пользовательскому рейтингу платформы PC`
alpha = 0.05
```

Для проверки гипопез будем использовать **t-критерий Стьюдента**, т.к. выборки не зависят друг от друга.

````python
data_selection['platform'].unique()
````

````python
#отфильтруем датафрейм по оценкам пользователей в разрезе платформ, удалим пропуски
data_Xbox = data_selection[data_selection['platform'] == 'XOne']['user_score'].dropna()
data_PC = data_selection[data_selection['platform'] == 'PC']['user_score'].dropna()

#проведем t-тест
results = st.ttest_ind(data_Xbox,data_PC, equal_var = False)

#задаем порог значимости
alpha = 0.05

print('p-значение:',results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
````

````python
#В доказательсто равенста вредних потребительских рейтингов на выбранных платформах посчитаем среднее
data_Xbox_mean = data_selection[data_selection['platform'] == 'XOne']['user_score'].dropna().mean()
data_PC_mean = data_selection[data_selection['platform'] == 'PC']['user_score'].dropna().mean()
print('Средний пользовательский рейтинг платформы Xbox One:{:.2f}'.format(data_Xbox_mean))
print('Средний пользовательский рейтинг платформы PС:{:.2f}'.format(data_PC_mean))
````

Не получилось отвергнуть гипотезу о том, что средние пользовательские рейтини платформ Xbox One и PC разные.

#### Проверка гипотезы: Средние пользовательские рейтинги жарнов Action и Sports разные

```
H_0: Средний пользовательский рейтинг жанра Action = среднему пользовательскому рейтингу жанра Sports`
H_1: Средний пользовательский рейтинг жанра Action ≠ среднему пользовательскому рейтингу жанра Sports`
alpha = 0.05
```

````python
#отфильтруем датафрейм по оценкам пользователей в разрезе жанров, удалим пропуски
data_action = data_selection[data_selection['genre'] == 'Action']['user_score'].dropna()
data_sports = data_selection[data_selection['genre'] == 'Sports']['user_score'].dropna()

#проведем t-тест
results = st.ttest_ind(data_action,data_sports, equal_var = False)

#задаем порог значимости
alpha = 0.05

print('p-значение:',results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
````

````python
#В доказательсто неравенства средних потребительских рейтингов в выбранных жанрах посчитаем среднее
data_action_mean = data_selection[data_selection['genre'] == 'Action']['user_score'].dropna().mean()
data_sports_mean = data_selection[data_selection['genre'] == 'Sports']['user_score'].dropna().mean()
print('Средний пользовательский рейтинг жанра Action: {:.2f}'.format(data_action_mean))
print('Средний пользовательский рейтинг жанра Sports:{:.2f}'.format(data_sports_mean))
````

На имеющихся данных, на уровне значимости 5% есть основания отвергнуть нулевую гипотезу о равенстве средних значений пользовательских рейтингов (отзывов) жанров Action и Sports в пользу альтернативной гипотезы.

Таким образом, мы принимаем альтернативную гипотезу, о неравенсте средних пользовательских рейтингов жанров Action и Sports.
